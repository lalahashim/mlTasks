```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("bank.xlsx")
```

```python
plt.figure(figsize=(20, 5))
sns.countplot(data=df, x='age', palette="flare")
plt.title("Customer's age")
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.countplot(data=df, x='job', palette="Set2")
plt.title("Customer's job")
plt.show()
```

```python
plt.figure(figsize=(25,15))
sns.boxplot(data=df, x='age', y='balance', palette='pastel')
plt.title("Balance distribution by age")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(data=df, x='job', y='balance',alpha=0.3, color='seagreen')
plt.title("Balance distribution by job")
plt.xlabel("job")
plt.ylabel("balance")
plt.show()
```

```python
plt.figure(figsize=(5, 5))

education_count = df['education'].value_counts()
colors = ["#59AC77", "#FDAAAA", "#FFF5E4", "#F0BB78"]

plt.pie(education_count, labels=education_count.index, autopct="%1.1f%%", colors=colors)
plt.title("Customer's education")
plt.show()
```
